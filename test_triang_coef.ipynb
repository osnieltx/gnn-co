```python
from graph import create_graph, triangle_coefficient, clustering_coefficient

import networkx as nx
import torch
```

```python
n, p = 10, .15
ei = create_graph(n, p)
edge_list = [(int(a), int(b)) for a, b in ei.T]
g_n = nx.from_edgelist(edge_list)
nx.draw(g_n, with_labels=True)
```

```python
cc = clustering_coefficient(ei)
tc = triangle_coefficient(ei, n, 8)
x = torch.cat((cc, tc), 1)
for i in range(n):
    print(f'{i=} {[f"{x_:.2f}" for x_ in x[i].tolist()]}')
```
